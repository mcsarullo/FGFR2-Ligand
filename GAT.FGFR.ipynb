```python
!pip install pandas numpy rdkit torch torch_geometric matplotlib
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
from IPython.display import display, Image
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GATv2Conv, global_mean_pool
from torch.nn import Linear, Dropout

# Data Loading and Preprocessing
def load_and_preprocess_data(filepath):
    """ loads csv data """
    try:
        df = pd.read_csv(filepath, sep=';', on_bad_lines='skip')
        print("First 5 rows:")
        display(df.head())
        print("Columns:")
        print(df.columns)
    except Exception as e:
        print("bad csv.")
        return pd.DataFrame()

    required_columns = ['Smiles', 'pChEMBL Value']
    if not all(col in df.columns for col in required_columns):
        print(f"the required columns are not in the df. make sure they are")
        print("available:", df.columns)
        return pd.DataFrame()


    df = df[required_columns].dropna()

    # pchembl to numeric to use as a pIC50
    df['pChEMBL Value'] = pd.to_numeric(df['pChEMBL Value'], errors='coerce')
    df = df.dropna(subset=['pChEMBL Value'])

    return df

# SMILES to a graph representation for GAT
def smiles_to_graph(smiles, y):
    """
    converts SMILES string to PyTorch Geometric graph, now including
    stereochemistry information (R/S for chiral centers).
    """
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None

    # Identify chiral centers and their stereochemistry
    Chem.AssignStereochemistry(mol, cleanIt=True, force=True)
    chiral_centers = Chem.FindMolChiralCenters(mol, force=True, includeUnassigned=False)
    chiral_dict = {center[0]: center[1] for center in chiral_centers}

    atom_features = []
    for i, atom in enumerate(mol.GetAtoms()):
        stereochemistry = 0
        if i in chiral_dict:
            # r is 1 and s is -1. nonchiral is 0
            stereochemistry = 1 if chiral_dict[i] == 'R' else -1

        atom_features.append([
            atom.GetAtomicNum(),
            atom.GetDegree(),
            atom.GetTotalNumHs(),
            atom.GetImplicitValence(),
            float(atom.GetIsAromatic()),
            float(stereochemistry)
        ])
    x = torch.tensor(atom_features, dtype=torch.float)

    edge_indices = []
    for bond in mol.GetBonds():
        i = bond.GetBeginAtomIdx()
        j = bond.GetEndAtomIdx()
        edge_indices.append((i, j))
        edge_indices.append((j, i)) # reverse direction since it's an undirected graph

    edge_index = torch.tensor(edge_indices, dtype=torch.long).t().contiguous()

    y_tensor = torch.tensor([y], dtype=torch.float)

    return Data(x=x, edge_index=edge_index, y=y_tensor, smiles=smiles)

# 2. GNN Model with Attention (GAT)
class GATModel(torch.nn.Module):
    """
    GAT for predicting molecular properties, prominently pIC50. Includes dropout as
    regularization and returns attention weights for visualization.
    """
    def __init__(self, num_node_features, hidden_channels, dropout_rate=0.5):
        super(GATModel, self).__init__()
        self.conv1 = GATv2Conv(num_node_features, hidden_channels, heads=4, concat=True, add_self_loops=False)
        self.dropout1 = Dropout(dropout_rate)
        self.conv2 = GATv2Conv(hidden_channels * 4, hidden_channels, heads=4, concat=True, add_self_loops=False)
        self.dropout2 = Dropout(dropout_rate)
        self.conv3 = GATv2Conv(hidden_channels * 4, hidden_channels, heads=4, concat=True, add_self_loops=False)
        self.dropout3 = Dropout(dropout_rate)
        self.conv4 = GATv2Conv(hidden_channels * 4, hidden_channels, heads=4, concat=True, add_self_loops=False)
        self.dropout4 = Dropout(dropout_rate)
        self.linear = Linear(hidden_channels * 4, 1)

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch

        # attention weights from first layer
        x, att_weights1 = self.conv1(x, edge_index, return_attention_weights=True)
        x = F.elu(x)
        x = self.dropout1(x)

        x = F.elu(self.conv2(x, edge_index))

        x = F.elu(self.conv3(x, edge_index))

        x = global_mean_pool(x, batch) # global mean representation

        x = self.linear(x)
        return x, att_weights1

# 3. Training and Evaluation Functions
def train(model, loader, optimizer, criterion, device):
    """training loop"""
    model.train()
    total_loss = 0
    for data in loader:
        data = data.to(device)
        optimizer.zero_grad()
        out, _ = model(data)
        loss = criterion(out, data.y.view(-1, 1))
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * data.num_graphs
    return total_loss / len(loader.dataset)

def evaluate(model, loader, device):
    """eval loop"""
    model.eval()
    total_squared_error = 0
    with torch.no_grad():
        for data in loader:
            data = data.to(device)
            out, _ = model(data)
            total_squared_error += F.mse_loss(out, data.y.view(-1, 1), reduction='sum').item()
    return total_squared_error / len(loader.dataset)

# 4. Attention Visualization
def visualize_attention(smiles, model, device, conv_layer_index=0):
    """put attention weights in a given prediction into a bar chart"""
    model.eval()
    graph = smiles_to_graph(smiles, 0)
    if graph is None:
        print("invalid smiles string")
        return

    data_loader = DataLoader([graph], batch_size=1)
    mol = Chem.MolFromSmiles(smiles)

    with torch.no_grad():
        for data in data_loader:
            data = data.to(device)
            _, att_weights = model(data)
            edge_index = att_weights[0].cpu().numpy()
            attention_values = att_weights[1].mean(dim=1).cpu().numpy()

            # dict for attention values for each bond
            bond_attention = {}
            for i in range(edge_index.shape[1]):
                u, v = edge_index[:, i]
                # average attn in both directions for an edge
                if (u, v) not in bond_attention and (v, u) not in bond_attention:
                    bond_attention[(u, v)] = attention_values[i]
                elif (u, v) in bond_attention:
                     bond_attention[(u, v)] = (bond_attention[(u,v)] + attention_values[i])/2
                else:
                    bond_attention[(v, u)] = (bond_attention[(v,u)] + attention_values[i])/2


            bond_labels = [f"{u}-{v}" for (u, v) in bond_attention.keys()]
            attention_scores = list(bond_attention.values())
            plt.figure(figsize=(12, 6))
            plt.bar(bond_labels, attention_scores)
            plt.xlabel("Bond (Atom Indices)")
            plt.ylabel("Attention Weight")
            plt.title(f"Attention Weights for Molecule: {smiles}")
            plt.xticks(rotation=90)
            plt.tight_layout()
            plt.show()


# main
if __name__ == '__main__':
    # config
    FILE_PATH = 'FGFR2_inhibitors.csv'
    BATCH_SIZE = 64
    LEARNING_RATE = 0.0001
    NUM_EPOCHS = 1000
    HIDDEN_CHANNELS = 128
    DROPOUT_RATE = 0.6

    # prep
    df = load_and_preprocess_data(FILE_PATH)
    data_list = [smiles_to_graph(smiles, y) for smiles, y in zip(df['Smiles'], df['pChEMBL Value'])]
    data_list = [data for data in data_list if data is not None]

    # split dataset
    torch.manual_seed(42)
    np.random.shuffle(data_list)
    train_size = int(0.8 * len(data_list))
    train_dataset = data_list[:train_size]
    test_dataset = data_list[train_size:]
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

    # initialize model
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    num_node_features = data_list[0].num_node_features
    model = GATModel(num_node_features=num_node_features, hidden_channels=HIDDEN_CHANNELS, dropout_rate=DROPOUT_RATE).to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)
    criterion = torch.nn.MSELoss()

    # training + eval
    print("Starting training...")
    for epoch in range(1, NUM_EPOCHS + 1):
        train_loss = train(model, train_loader, optimizer, criterion, device)
        test_mse = evaluate(model, test_loader, device)
        if epoch % 25 == 0: # print every 25 epoch
            print(f'Epoch: {epoch:03d}, Train Loss: {train_loss:.4f}, Test MSE: {test_mse:.4f}')

    print("Training finished.")

    # final eval
    final_test_mse = evaluate(model, test_loader, device)
    print(f"\nFinal Test MSE: {final_test_mse:.4f}")

    # one prediction as an example
    def predict_pchembl(smiles_string, trained_model, device):
        """predicts pIC550"""
        trained_model.eval()
        graph = smiles_to_graph(smiles_string, 0)
        if graph is None:
            return "Invalid SMILES string"

        data_loader = DataLoader([graph], batch_size=1)
        with torch.no_grad():
            for d in data_loader:
                d = d.to(device)
                prediction, _ = trained_model(d)
        return prediction.item()

    # e.g. from dataset
    new_smiles = 'COc1cccc(Nc2ncc3cc(-c4c(Cl)cccc4Cl)c(=O)n(C)c3n2)c1'
    predicted_pchembl = predict_pchembl(new_smiles, model, device)
    print(f"Predicted pChEMBL value for '{new_smiles}': {predicted_pchembl:.4f}")

    # visualize attn weights
    print(f"\nVisualizing attention for '{new_smiles}':")
    visualize_attention(new_smiles, model, device)
```
